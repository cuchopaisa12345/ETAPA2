```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv('cars.csv')

df.head(10)




# grafica de dispersion

df.plot.scatter(x="year",y="priceUSD")

plt.show

#regresion lienal
regresion= linear_model.LinearRegression()

#agrego datos al vector

year=df["year"].values.reshape((-1,1))

#Se crea el modelo
modelo=regresion.fit(year, df["priceUSD"])

print ("interseccion.(b)",modelo.intercept_)

#Imprimimos pendiente 

print ("Pendiente.(m)", modelo.coef_)

entrada=[[1955],[1988],[1999],[2000],[ 2012]]

predicciones = modelo.predict(entrada)
print("las predicciones son:", predicciones)

df.plot.scatter(x="year",y="priceUSD",label="Datos originales")
plt.scatter(entrada,predicciones,color="red",label ="Predicciones")
plt.plot(entrada,predicciones,color="black",label="Linea de regresion")

plt.xlabel("year")
plt.ylabel("price")
plt.legend()
plt.show()

```
